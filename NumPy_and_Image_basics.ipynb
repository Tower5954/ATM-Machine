# What is an image?

In computer vision, every image can be represented as an array.

 This is an example of how a computer sees a single digit in a 28 x 28 pixel image.

 /content/img_to_array.jpg

 Often the default images have values between 0 and 255.
We can 'normalise' numbers by 255 to represent them as between 0 or 1.


Colour is represented as a combination of Red, Blue and Green (RGB).
So the shape of a colour array has 3 dimensions.
* Height
* Width
* Colour channels

This means when you read into an image and then check its shape, it will look something like:    
**(1280,780,3)**.
* 1280 pixel height
* 780 pixel width
* 3 colour channels
